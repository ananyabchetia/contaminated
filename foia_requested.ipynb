## Open the xlsx file

```python
import pandas as pd
df = pd.read_excel('JH Baxter Complaints 9-12-1977 thru 09-28-2022 with Confidential blocked.xlsx')
df
```

### Making a separate column for year 

```python
df['Date of Complaint Submission'] = pd.to_datetime(df['Date of Complaint Submission'])
df['Year'] = df['Date of Complaint Submission'].dt.year
df
```

```python
df['Complainant City'].value_counts()
```

```python
df_1=df['Year'].value_counts().reset_index()
```

```python
df_1['count'].sum()
```

```python
year_counts = df['Year'].value_counts().reset_index()
year_counts.columns = ['Year', 'Count']
year_counts = year_counts.sort_values(by='Year')  
year_counts
```

```python
year_counts.to_csv('foia_years_3.csv')
```
